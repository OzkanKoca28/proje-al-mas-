```python
# PROJE 4 : Telefon Rehberi

kişiler = {}      #  Liste "[]" ile , Dict(sözlük) "{}" ile kullanılır..

def menü():
    print("---TELEFON REHBERİ MENÜSÜ---")
    print("1. Kişi Ekle")
    print("2. Kişi Ara")
    print("3. Kişileri Listele")
    print("4. Kişi Sil")
    print("5. Telefon Numarası Güncelle")
    print("6. Toplam Kişi Sayısı")
    print("0. Çıkış Yap")



def kişi_ekle():

    isim = input("Kayıt etmek istediğiniz kişinin ismini giriniz...").capitalize()

    if any(harf.isdigit() for harf in isim):
        print("İsim rakam içeremez.")

    else:
        if isim in kişiler:
            print(f"{isim} kişisi rehberinizde kayıtlı")
            seçim = input(f"{isim} kişisinin numarası değiştirilsin mi? (EVET/HAYIR) ").upper()

            if seçim == "EVET":
                numara = input("Yeni numarayı giriniz: ")
                numara = numara.replace(" ", "")

                if not numara.isdigit():
                    print("Numara sadece rakamlardan oluşmalıdır.")

                elif len(numara) != 10:
                    print("Lütfen 10 haneli numara giriniz...")

                else:
                    kişiler[isim] = numara
                    print(f"{isim} kişisinin numarası {numara} olarak değiştirildi.")

            else:
                print("İşlem iptal edildi.")

        else:
            numara = input("Lütfen kayıt etmek istediğiniz numarayı giriniz: ")
            numara = numara.replace(" ", "")

            if not numara.isdigit():
                print("Numara sadece rakamlardan oluşmalıdır.")

            elif len(numara) != 10:
                print("Lütfen 10 haneli bir numara giriniz...")

            else:
                kişiler[isim] = numara
                print(f"{isim} kişisi {numara} numarası ile rehberinize başarıyla eklendi.")



def kişi_ara():
        isim = input("Aramak istediğiniz kişinin ismini giriniz...").capitalize()

        if isim in kişiler:
            print("Kişi bulundu")
            print(f"{isim}: {kişiler[isim]}")

        else:
            print(f"{isim} kişisi kayıtlı değil.")




def kişileri_listele():
    if not kişiler:
        print("Kişi rehberiniz boş")

    else:
        print("--Kişi Rehberi--")
        for isim,numara in kişiler.items():
            print(f"{isim}: {numara}")



def kişi_sil():
        isim = input("Silmek istediğiniz kişinin ismini giriniz...").capitalize()

        if isim in kişiler:
            del kişiler[isim]
            print(f"{isim} kişisi başarıyla silindi.")

        else:
            print(f"{isim} kişisi rehberinizde mevcut değil.")



def telefon_numarası_güncelle():    

    isim = input("Lütfen numarasını güncellemek istediğiniz kişinin ismini giriniz...").capitalize()

    if isim not in kişiler:
        print(f"{isim} kişisi kayıtlı değil.")

    else:
        numara = input("Lütfen güncellemek istediğiniz numarayı giriniz...")
        numara = numara.replace(" ","")

        if not numara.isdigit():
            print("Numara sadece rakamlardan oluşmalıdır...")

        elif len(numara) != 10 :
            print("Lütfen 10 haneli bir numara giriniz...")

        else:
            kişiler[isim] = numara
            print(f"{isim} kişisinin numarası {numara} numarası ile güncellendi.")





def toplam_kişi_sayısı():
    if not kişiler:
        print("Rehberinizde kayıtlı kişi bulunmamaktadır.")

    else:
        print(f"Toplam kişi sayısı: {len(kişiler)}")


def çıkış_yap():
    print("Rehberden çıkış yapılıyor...")




while True:

    menü()
    seçim = input("Lütfen menüden yapmak istediğiniz işlemin numarasını giriniz ")

    if seçim == "1":
        kişi_ekle()

    elif seçim == "2":
        kişi_ara()

    elif seçim == "3":
        kişileri_listele()

    elif seçim == "4":
        kişi_sil()

    elif seçim == "5":
        telefon_numarası_güncelle()

    elif seçim == "6":
        toplam_kişi_sayısı()

    elif seçim == "0":
        çıkış_yap()
        break
    
    else:
        print("Hatalı numara, Lütfen menüdeki işlem numaralarından giriniz...")
```
