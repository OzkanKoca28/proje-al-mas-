```python
# PROJE 2 : ATM SİSTEMİ

bakiye = 1000

def menu():
    print("---ATM SYSTEM MENU---")
    print("1. Bakiye Göster")
    print("2. Para Yatır")
    print("3. Para Çek")
    print("4. Çıkış Yap")


def bakiye_göster():
    print(f"{bakiye} Güncel bakiyeniz")

def para_yatır():
    global bakiye

    try:
        para = int(input("Lütfen yatırmak istediğiniz para miktarı giriniz"))

        if para <= 0 :
            print("Lütfen geçerli miktar giriniz...")

        else:
            bakiye += para
            print(f"{para} TL Miktarında para başarıyla bakiyenize eklenmiştir.")
    
    except ValueError:
        print("Lütfen sadece sayı giriniz...")



def para_çek():
    global bakiye
    
    try:
        para = int(input("Lütfen çekmek istediğiniz para miktarını giriniz"))

        if para <= 0:
            print("Lütfen geçerli miktar giriniz...")

        else:
            if bakiye < para :
                print("Yeterli bakiyeniz yok.")

            else:
                bakiye -= para
                print(f"{para} TL Miktarında para başarıyla hesabınızdan çekilmiştir. ")

    except ValueError:
        print("Lütfen sadece sayı giriniz...")


def çıkış_yap():
    print("Sistemden çıkış yapılıyor...")



while True:
    
    menu()

    seçim = input("Lütfen yapmak istediğiniz işlemin menüdeki işlem numarasını giriniz...")

    if seçim == "1":
        bakiye_göster()

    elif seçim == "2":
        para_yatır()

    elif seçim == "3":
        para_çek()

    elif seçim == "4":
        çıkış_yap()
        break

    else:
        print("Hatalı numara, Lütfen menüdeki numaralardan seçiniz (1-4)")

```
