```python
# PROJE 1 : TO DO LİST (YAPILACAKLAR LİSTESİ)

# Değişken, programın başında görev(yapılacaklar) listemiz boş
görevler = []

def menü():
    print("---TO DO LİST MENU")
    print("1. Görev Ekle")
    print("2. Görev Sil")
    print("3. Görevleri Listele")
    print("4. Görev Ara")
    print("0. Programdan Çık")


def görev_ekle():
        görev = input("Eklemek istediğiniz görevi yazınız").capitalize()
        görevler.append(görev)
        print(f"{görev} görevi listeye başarıyla eklendi.")


def görev_sil():
    if len(görevler) == 0 :
        print("Listeniz boş, Silinecek görev yok.")

    else:
        görev = input("Silmek istediğiniz görevi yazınız.").capitalize()
        if görev in görevler: 
            görevler.remove(görev)
            print(f"{görev} görevi listeden başarıyla silindi.")

        else:
            print("Silmek istediğiniz görev listede yok.")


def görevleri_listele():
    if len(görevler) == 0 :
       print("Görev listeniz boş.")

    else:
        for i,görev in enumerate(görevler , 1):
            print(f"{i}. {görev}")



def görev_ara():
        görev = input("Aramak istediğiniz görevi yazınız").capitalize()
        if görev in görevler:
            print(f"{görev} listede bulundu.")
        
        else:
            print(f"{görev} listede bulunamadı.")

def çıkış_yap():
    print("Programdan çıkış yapılıyor...")



while True:  # While sonsuz döngüdür. While bloğu True olduğu sürece tekrar tekrar çalışır, break fonksiyonu ile sonlandırabiliriz.

    menü()
  
    seçim = input("Lütfen yapmak istediğiniz işlemin menü'de ki numarasını giriniz... ")

    if seçim == "1":
        görev_ekle()

   
    elif seçim == "2":
        görev_sil()

    
    elif seçim == "3":
        görevleri_listele()

   
    elif seçim == "4":
        görev_ara()
  

    elif seçim == "0":
        çıkış_yap()
        break

    else:
        print("Hatalı numara, Lütfen menüde belirtilen 1,2,3,4,0 numaralarından seçiniz.")


```
