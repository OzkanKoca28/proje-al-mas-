```python
# PROJE 3 : Öğrenci Not Sistemi

öğrenciler = {}

def menü():
    print("--- Öğrenci Not Sistemi ---")
    print("1. Öğrenci Ekle")
    print("2. Not Güncelle")
    print("3. Öğrencileri Listele")
    print("4. Öğrenci Ara")
    print("5. Öğrenci Sil")
    print("6. Toplam Öğrenci Sayısı")
    print("0. Çıkış Yap")



def öğrenci_ekle():
    try:
        isim = input("Öğrencinin adını giriniz...").capitalize()

        if isim in öğrenciler:
            print(f"{isim} isimli öğrenci zaten kayıtlı.")

        else:
            notu = int(input("Öğrencinin notunu giriniz..."))

            if notu < 0 or notu > 100:
                print("Not 0 ile 100 arasında olmalı.")

            else:
                öğrenciler[isim] = notu
                print(f"{isim} isimli öğrenci {notu} notuyla başarıyla eklendi.")

    except ValueError:
        print("Lütfen notu sadece sayı olarak giriniz...")



def not_güncelle():
    try:
        isim = input("Güncellemek istediğiniz öğrencinin ismini giriniz...").capitalize()

        if isim in öğrenciler:

            notu = int(input("Güncellemek istediğiniz notu giriniz..."))

            if notu < 0 or notu > 100:
                print("Girilen not 0 ile 100 arasında olmalı.")

            else:
                öğrenciler[isim] = notu
                print(f"{isim} isimli öğrencinin notu {notu} olarak başarıyla güncellendi.")

        else:
            print(f"{isim} isimli öğrenci bulunmamaktadır.")

    except ValueError:
        print("Lütfen notu sadece sayı olarak giriniz...")

    

def öğrencileri_listele():

    if not öğrenciler :
        print("Listeniz boş")

    else:
        print("-- Öğrenci Listesi --")
        for öğrenci,notu in öğrenciler.items():
            print(f"{öğrenci}: {notu}")



def öğrenci_ara():
    if len(öğrenciler) == 0 :
        print("Listeniz boş")

    else:
        öğrenci = input("Aramak istediğiniz öğrencinin ismini giriniz...").capitalize()

        if öğrenci in öğrenciler:
            print(f"{öğrenci}: {öğrenciler[öğrenci]}")

        else:
            print(f"{öğrenci} isimli öğrenci bulunamadı.")



def öğrenci_sil():
    if not öğrenciler :
        print("Kayıtlı öğrenci bulunmamaktadır.")

    else:
        öğrenci = input("Silmek istediğiniz öğrencinin ismini giriniz...").capitalize()

        if öğrenci in öğrenciler:
            del öğrenciler[öğrenci]
            print(f"{öğrenci} isimli öğrencisi başarıyla silindi.")

        else:
            print(f"{öğrenci} isimli öğrenci listenizde zaten yok.")



def toplam_öğrenci():
    if not öğrenciler:
        print("Kayıtlı öğrenci bulunmamaktadır.")

    else:
        print(f"Toplam Öğrenci Sayısı: {len(öğrenciler)}")

    
    
def çıkış_yap():
    print("Programdan çıkış yapılıyor...")



while True:

    menü()

    seçim = input("Lütfen seçtiğiniz işlemin menüdeki numarasını giriniz...")

    if seçim == "1":
        öğrenci_ekle()

    elif seçim == "2":
        not_güncelle()

    elif seçim == "3":
        öğrencileri_listele()

    elif seçim == "4":
        öğrenci_ara()

    elif seçim == "5":
        öğrenci_sil()

    elif seçim == "6":
        toplam_öğrenci()

    elif seçim == "0":
        çıkış_yap()
        break

    else:
        print("Hatalı numara, Lütfen menüdeki işlem numaralarından seçim yapınız...")



```
